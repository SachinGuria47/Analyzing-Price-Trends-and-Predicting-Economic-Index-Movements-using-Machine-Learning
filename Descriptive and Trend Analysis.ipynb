# #Analyzing Price Trends and Inflation Patterns in India's Consumer Price Index (Rural and Urban) up to February 2024


```python
#import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#Load Data

```python
df = pd.read_csv("cleaned_cpi.csv")
df.info()  #checking the datatypes and value counts in each columns
```

# Descriptive Analysis ðŸ‘‡


*   ## Summary Statistics




```python
df.describe()
```

# Distribution of the numerical features


```python
num_col = df.select_dtypes(include=np.number).columns.tolist() # to filter out only the columns that contain numerical data types

num_col.remove('Year')
fig,axes = plt.subplots(6,5,figsize=(20,20)) #6 rows and 5 columns
axes = axes.flatten()
for i, col in enumerate(num_col):
    df[col].plot(kind='hist',bins=15, ax=axes[i], edgecolor='black', title=col)
plt.subplots_adjust(hspace=0.5)
plt.show()
```

# Boxplot for identifying outliers




```python
plt.figure(figsize=(15,10))
sns.boxplot(data=df[num_col],orient='h')
plt.xticks(rotation=0)
plt.title('Boxplot for identifying outliers')
plt.show()
```

# yearly trends visualization

```python
numeric_columns = df.select_dtypes(include=np.number).drop(columns=['Year'])
yearly_trends = numeric_columns.groupby(df['Year']).mean()
for category in yearly_trends.columns:
    plt.plot(yearly_trends.index, yearly_trends[category], marker='o', linestyle='-', label=category)

plt.xlabel('Year')
plt.ylabel('CPI')
plt.title('Yearly Trends in CPI')
plt.legend(loc='upper right', bbox_to_anchor=(2, 1))
plt.grid(True)
plt.show()
```

# correlation analysis

```python
numeric_columns = df.select_dtypes(include=np.number).drop(columns=['Year'])
correlation = numeric_columns.corr()

# Create a heatmap of the correlation matrix
mask = np.triu(np.ones_like(correlation, dtype=bool))
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, mask=mask,annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 8})
plt.title('Correlation Matrix')
plt.show()
```

## Removing highly correlated values ranging between > 0.94 and <1

```python
correl = numeric_columns.corr().abs()

to_drop = set()
for i in range(len(correl.columns)):
    for j in range(i):
        if abs(correl.iloc[i, j]) > 0.94 and abs(correl.iloc[i, j])< 1:
            colname = correl.columns[i]
            to_drop.add(colname)

reduced_data = numeric_columns.drop(columns=to_drop)
reduced_data.info()
```

## New Correlation ðŸ‘‡

```python
cor = reduced_data.corr()
mask = np.triu(np.ones_like(cor, dtype=bool))
plt.figure(figsize=(12, 10))
sns.heatmap(cor, mask=mask,annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 8})
plt.title('Correlation Matrix')
plt.show()
```

# Trend Analysis
## Overall CPI Trend Analysis of various features over the Years  ðŸ‘‡

```python
yearly_trends = numeric_columns.groupby(df['Year']).mean()

for category in reduced_data.columns:
    plt.plot(yearly_trends.index, yearly_trends[category], marker='o', linestyle='-', label=category)

plt.xlabel('Year')
plt.ylabel('CPI')
plt.title('Yearly Trends in CPI')
plt.legend(loc='upper right', bbox_to_anchor=(2, 1))
plt.grid(True)
plt.show()
```

## CPI Trends for Rural and Urban Areas over the years ðŸ‘‡`

```python
# filtering and calculating the mean for rural and urban sector
rural = reduced_data[df['Sector'] == 'Rural'].groupby(df['Year']).mean()
urban = reduced_data[df['Sector'] == 'Urban'].groupby(df['Year']).mean()

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
for category in reduced_data.columns:
    plt.plot(rural.index, rural[category], marker='o', linestyle='-', label=category)

plt.xlabel('Year')
plt.ylabel('CPI')
plt.title('Yearly Trends in CPI for Rural Areas')
plt.legend(fontsize='x-small', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)


plt.subplot(1, 2, 2)
for category in reduced_data.columns:
    plt.plot(urban.index, urban[category], marker='o', linestyle='-', label=category)

plt.xlabel('Year')
plt.ylabel('CPI')
plt.title('Yearly Trends in CPI for Urban Areas')
plt.legend(fontsize='x-small', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)


plt.tight_layout()
plt.show()
```
